```python
import pandas as pd

data = pd.read_csv("C:/Users/tanis/Desktop/Celebal/Superstore-Sales.csv", encoding='latin1')

print(data.info())
print(data.head())
```

```python
print(data.head())
```

```python
df=pd.DataFrame(data)
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


df.columns = df.columns.str.strip()

plt.figure(figsize=(10,6))
sns.barplot(x='Product Category', y='Sales', data=df, estimator=sum)
plt.title('Total Sales by Product Category')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python

plt.figure(figsize=(12,6))
sns.barplot(x='Product Sub-Category', y='Profit', data=df, estimator=sum)
plt.title('Total Profit by Product Sub-Category')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='Sales', y='Profit', hue='Product Category')
plt.title('Sales vs Profit by Product Category')
plt.tight_layout()
plt.show()
```

```python
top_provinces = df.groupby('Province')['Sales'].sum().nlargest(10).reset_index()

plt.figure(figsize=(10,6))
sns.barplot(data=top_provinces, x='Province', y='Sales')
plt.title('Top 10 Provinces by Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
